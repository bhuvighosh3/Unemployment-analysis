```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Load the data into a pandas DataFrame

# Define a function to calculate the MAPE
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100



# Define a function to calculate the ARIMA forecast
def arima_forecast(df, p=1, d=1, q=1):
    # Create a new DataFrame with the same columns as the input DataFrame
    df_arima = pd.DataFrame(columns=df.columns)

    # Loop over all the countries in the input DataFrame
    for country_code in df['Country_code'].unique():
        # Select only the rows that correspond to the current country code
        df_country = df.loc[df['Country_code'] == country_code]

        # Convert the DataFrame to a time series
        ts = pd.Series(df_country['No. of unemployed'].values, index=pd.to_datetime(df_country['Year'], format='%Y'))

        # Fit an ARIMA model to the time series
        model = ARIMA(ts, order=(p, d, q))
        results = model.fit()

        # Use the ARIMA model to forecast the next 2 years
        forecast = results.forecast(steps=2)

        # Add a new row to the output DataFrame with the forecasted values
        df_arima = df_arima.append({'Country_code': country_code, 'Year': 2022, 'No. of unemployed': forecast[0]},
                                   ignore_index=True)
        df_arima = df_arima.append({'Country_code': country_code, 'Year': 2023, 'No. of unemployed': forecast[1]},
                                   ignore_index=True)

    return df_arima

```

```python
def calculate_pct_change(df, country_code):
    # Select only the rows that correspond to the specified country code
    df_country = df.loc[df['Country_code'] == country_code]

    # Calculate the difference between each element and its previous element
    diff = df_country['No. of unemployed'].diff()

    # Calculate the percentage change by dividing the difference by the previous element and multiplying by 100
    pct_change = diff / df_country['No. of unemployed'].shift() * 100

    # Return the result
    return pct_change
```
